```python
import numpy as np
import glob
import nibabel as nib
import random
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import matplotlib.pyplot as plt
from nilearn import plotting
from nilearn import image
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
miccai_path = os.getenv('MICCAI_DATA')
train_path = f"{miccai_path}/training-images/"
files = glob.glob(f"{train_path}*.nii.gz")
```

```python
file_to_vis = files[1]
img = nib.load(file_to_vis).get_fdata()
img.shape
```

```python
@interact
def plot_img(volume=[i for i in range(1, img.shape[3] + 1)], slice=[i for i in range(1, img.shape[2])]):
    plt.imshow(img[:, :, slice-1, volume-1], cmap='gray')
```

```python
plotting.plot_img(image.index_img(files[1], 0))
plotting.plot_img(image.index_img(files[1], 1))
```
